В таблице приведены данные о взаимонезависимых доходностях трех акций: A, B, C.

Момент наблюдения	Доходность A, %	Доходность B, %	Доходность C, %
1	25	0	10
2	-10	15	25
3	10	-5	-15
4	5	5	-5
5	35	20	-5
6	13	25	15
Постройте линии регрессии, с помощью функции regplot, зависимости доходности одной акции от другой. Всего три графика. Выберите пару акций которые наименее зависимы в доходностях от друг друга.

```python
import pandas as pd
import seaborn as sns
link= "https://goit.global/textbooks/python/data-science-xf8b2g/v1/docs/module-03/hw/"
data = pd.read_html(link)
data=data[0]
data.columns
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(data['Доходность A, %'],data['Доходность B, %'])

print(r_value*r_value)

sns.set_theme(color_codes=True)
plt.figure(figsize = (10,10))
ax=sns.regplot(x='Доходность A, %', y='Доходность B, %',   data=data,  x_estimator=np.median)
```

***Данная пара наименее зависимы в доходностях от друг друга***

```python
ax1=sns.regplot(x='Доходность A, %', y='Доходность C, %',  data=data, x_estimator=np.median)

slope, intercept, r_value, p_value, std_err = stats.linregress(data['Доходность A, %'],data['Доходность C, %'])

print(r_value*r_value)
```
